```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
stocks = ["RELIANCE.NS","HDFCBANK.NS","ICICIBANK.NS","INFY.NS","ITC.NS",
                "TCS.NS","AXISBANK.NS","KOTAKBANK.NS","HINDUNILVR.NS",
                "SBIN.NS","BHARTIARTL.NS","BAJFINANCE.NS","ASIANPAINT.NS","M&M.NS",
                "MARUTI.NS","HCLTECH.NS","SUNPHARMA.NS","LTIM.NS","TATAMOTORS.NS","ULTRACEMCO.NS"]
```

```python
data = yf.download(stocks, start="2018-01-01", end="2023-01-01")['Adj Close']
```

```python
returns = data.pct_change().dropna()
```

```python
correlation_matrix = returns.corr()
```

```python
correlation_matrix
```

```python
top_pairs = correlation_matrix.unstack().sort_values(ascending=True)
```

```python
top_pairs
```

```python
top_pairs=top_pairs.drop_duplicates()
```

```python
top_pairs=top_pairs.head(20)
```

```python
top_pairs
```

```python
reqlist=[]
for  pair in top_pairs.index:
    stock1, stock2 = pair
    reqlist.append(stock1)
    reqlist.append(stock2)
```

```python
reqlist_df = pd.DataFrame(reqlist, columns=['Stock'])
reqlist_df = reqlist_df.drop_duplicates(subset='Stock')
reqlist_df.set_index('Stock', inplace=True)
reqlist_df=reqlist_df.head(10)
reqlist_df
```

```python
means_list = []
for stock in reqlist_df.index:
    mean_stock = correlation_matrix.loc[stock, reqlist_df.index].mean()
    means_list.append((stock, mean_stock))  
```

```python
means_list
```

```python
df = pd.DataFrame(means_list, columns=['Stock', 'Mean'])
```

```python
df = df.drop_duplicates(subset='Stock').sort_values(by='Mean')
df.set_index('Stock', inplace=True)
```
